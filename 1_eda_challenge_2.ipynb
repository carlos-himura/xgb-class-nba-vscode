Conclusions
1. There are 4 features to one-hot encoding (position,	team,	opponent,	game_location)
2. We need to remove row_id, and player_id as they are not adding any relevant information
- row_id is just an index, a sequential number that identifies each row. It has no relationship to the target
- player_id is a unique identifier, not a performance feature. Then the model is not learning basketball patterns — it’s memorizing players, not performance factors.
3. No duplicated rows and no missing info 
4. There are no inconsistencies in the categorial and numerical values

```python
import pandas as pd

# Define the columns to be removed
columns_to_drop = ['player_id', 'row_id'] 

# Drop the columns. The axis=1 argument specifies that we are dropping columns, not rows.
# errors='ignore' ensures the code doesn't crash if 'row_id' is not found.
df_clean = df_initial.drop(columns=columns_to_drop, axis=1, errors='ignore')

print("✅ Columns dropped successfully.")
print(f"Old shape: {df_initial.shape}")
print(f"New shape: {df_clean.shape}")

df_clean.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = [
    'age', 'minutes_played', 'points', 'rebounds', 'assists',
    'steals', 'blocks', 'turnovers', 'fg_pct', 'three_pct',
    'ft_pct', 'plus_minus', 'efficiency', 'rest_days'
]

plt.figure(figsize=(20, 15))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(5, 3, i)
    sns.boxplot(x=df_initial[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

Example interpretation for age:
- The age distribution is centered around 26–27 years. Most players fall between 24 and 29 (IQR). The majority range is between 20 and 35 years old. A few players aged 37–39 appear as outliers, but these are valid extreme values and not errors

- FG% → overall scoring efficiency
- 3PT% → long-range shooting effectiveness
- FT% → ability to capitalize on foul shots

- plus_minus: How much did the team outscore or get outscored by while this player was playing?. Plus/Minus shows impact on team success, not just individual stats.

```python
# Correlation analysis to check correlations between numerical features and multicollinearity
corr = df_clean.corr(numeric_only=True)
plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=False, cmap="coolwarm")
```

```python
df_clean.corr(numeric_only=True)['target'].sort_values(ascending=False)
```

Correlation Analysis Summary:
1. Strongly correlated features (best predictors): efficiency (0.68), minutes_played (0.55), points (0.54), rebounds (0.39), and assists (0.33).
- top predictive signals for the model.
2. Moderately correlated features: steals (0.26), blocks (0.23), plus_minus (0.15), and turnovers (0.13).
- These add useful but smaller predictive power.
- They help tree-based models because they capture complementary performance aspects (Nonlinear interactions).
3. Very weak or near zero correlation: fg_pct (0.008), age (0.005), rest_days (–0.006), three_pct (–0.01), and ft_pct (–0.02).
- Individually, these features do not correlate with the target in a linear way. But that does not mean they are useles.
- They may not correlate linearly with the target, but could interact with other features

Action: Do NOT drop them yet
- Tree models (RandomForest/XGBoost) can use nonlinear patterns.


Multicollinearity analysis
1. There is high multicollinearity present, particularly between efficiency and several other statistics (e.g., points, plus_minus).

2. For Tree-Based Models: ignore the high multicollinearity. Tree ensembles are robust to this, as they look at one feature at a time to determine splits. If two features contain the same information (redundant), the tree simply: picks the most useful one at each split, and ignores the redundant one. 

3. For Linear Models (Logistic Regression Baseline) multicollinearity causes SERIOUS problems: unstable and difficult-to-interpret model coefficients. It should be address:
- Feature Removal: Drop one of the highly correlated features. 
- Regularization: Use L2 (Ridge) or L1 (Lasso) regularization in your Logistic Regression model. These techniques penalize large coefficients and help stabilize the model in the presence of collinearity.

Action:
- As the goal is predictive performance, better to use tree-based models (XGBoost, Random Forest, LightGBM, etc). 
- Keep all features (trees will decide importance).

```python
# One-Hot Encoding for Categorical Variables

# Assuming your cleaned data is in df_clean
categorical_features = ["position", "team", "opponent", "game_location"]

df_encoded = pd.get_dummies(
    df_clean,
    columns=categorical_features,
    drop_first=True  # avoids dummy variable trap
)

# Convert all boolean columns to integers
df_preprocessed = df_encoded.astype({col: "int" for col in df_encoded.select_dtypes("bool").columns})

df_preprocessed.head()
```
